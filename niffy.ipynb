```python
nifty50["Tomorrow"] = nifty50["Close"].shift(-1)
```

```python
nifty50["Target"]= (nifty50["Tomorrow"] > nifty50["Close"]).astype(int)
```

```python
nifty50 = nifty50.loc["2006-01-01":].copy()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
train = nifty50.iloc[:-100]
test = nifty50.iloc[-100:]
predictors  =  ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

pred  = model.predict(test[predictors])
```

```python
import pandas as pd
pred = pd.Series(pred, index=test.index)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    pred = model.predict(test[predictors])
    pred = pd.Series(pred, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], pred], axis=1)  # Corrected 'axix' to 'axis'
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)  # Corrected 'tarin' to 'train'
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(nifty50, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = nifty50.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    nifty50[ratio_column]= nifty50["Close"] / rolling_averages["Close"]
    trend_column = f"Trend_{horizon}"
    
    nifty50[trend_column] = nifty50.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
nifty50 = nifty50.dropna()
```

```python
model = RandomForestClassifier(n_estimators= 200, min_samples_split=50, random_state=1 )
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    pred = model.predict_proba(test[predictors])[:,1]
    pred[pred >= .6] =1
    pred[pred < .6] =0
    pred = pd.Series(pred, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], pred], axis=1)  # Corrected 'axix' to 'axis'
    return combined
```

```python
predictions = backtest(nifty50, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
